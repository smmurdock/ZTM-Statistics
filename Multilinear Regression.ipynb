```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
df = pd.read_csv('diamond.csv')
df.head()
```

```python
# DF information
df.info()
```

# Exploratory Data Analysis

```python
# Correlation
df[['carat', 'price']].corr()
```

```python
# Isolate X and Y
y = df.price
X = df.carat
```

```python
# Scatter plot
plt.scatter(X, y)
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Relationship of Carat vs Price')

plt.show();
```

# Linear Regression

Equation: y = a + (b * X) + e

```python
# Add a constant
# This is a redundant step but necessary
X = sm.add_constant(X)
X.head(2)
```

```python
model1 = sm.OLS(endog = y, exog = X).fit()
print(model1.summary())
```
